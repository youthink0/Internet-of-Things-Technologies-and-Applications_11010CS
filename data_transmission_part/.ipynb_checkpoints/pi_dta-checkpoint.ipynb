```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt

day_in_now = datetime.date.today()
day_in_21ago = day_in_now + datetime.timedelta(days=-21)
def export_line_chart(df, df_week, today):
    #line chart
    plt.style.use("ggplot")               # 使用ggplot主題樣式

    #畫第一條線，plt.plot(x, y, c)參數分別為x軸資料、y軸資料及線顏色 = 紅色
    plt.plot(df["P"], df["P"], 's-', c = "r", label="today")
    plt.plot(df["P"], df_week["P"], 'o-', c = "g", label="week ago")

    # 設定圖例，參數為標籤、位置
    plt.legend(loc = 'best')
    plt.xlabel("count", fontweight = "bold")                # 設定x軸標題及粗體
    plt.ylabel("value", fontweight = "bold")    # 設定y軸標題及粗體
    plt.title("example", fontsize = 15, fontweight = "bold", y = 1.1)   # 設定標題、文字大小、粗體及位置
    plt.xticks(rotation=45)   # 將x軸數字旋轉45度，避免文字重疊

    plt.savefig("picture/"+str(today)+"_example.jpg",   # 儲存圖檔
                bbox_inches='tight',               # 去除座標軸占用的空間
                pad_inches=0.0)                    # 去除所有白邊
    plt.close()      # 關閉圖表

# right method 
with pd.ExcelFile('data_storage/pandas_simple.xlsx') as xls:
    for sheet_name in xls.sheet_names:
        df = pd.read_excel(xls, sheet_name=sheet_name)
        
        month_tmp = datetime.datetime.strptime(df["month"][0], "%b")
        sheet_date = str(df['year'][0])+"-"+str(month_tmp.month)+"-"+str(df['date'][0])
        #print(sheet_date)
        
        tmp = datetime.datetime.strptime(sheet_date, "%Y-%m-%d")
        week_ago_date = (tmp + datetime.timedelta(days=1)).strftime("%a %b %d")
        week_ago_date = week_ago_date.replace("0", " ")
        #print(week_ago_date)
        
        df_week_ago = pd.DataFrame()
        if week_ago_date in xls.sheet_names:
            df_week_ago = pd.read_excel(xls, sheet_name=week_ago_date)
            print(df_week_ago.head(2))
            
        export_line_chart(df, df_week_ago, sheet_date)
        
        print(df.head(3))
        print("------------------------------------------------------------")
```
